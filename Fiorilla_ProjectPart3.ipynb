## Section 1: Data and Research Questions
#### Data
https://www.nature.com/articles/s41597-023-02836-2 <p>
In the 2024 LaFrance et. al. paper, the authors compile and look at amphibian diversity across all U.S. National Park Services (NPS) managed land. In this case, <b>each documented amphibian species is an observation</b>. Starting with a 2021 dataset managed by the NPS called NPSpecies, there were 4,198 species. The authors validated the presence of amphibian species at each park by consulting primary literature, government reports, land guides, and verified iNaturalist submissions, and regional experts. This process resulted in the removal of 836 species and the addition of 115 species, resulting in a total of 3,472 species. <p>
For each species, they also updated and added features including <b>park identifiers</b> (name, code, network, region, state), <b>taxonomic classifications</b> (serial number, order, family, scientific name, common name), <b>occurrence record</b> (date, location within park, verification source, notes), <b>conservation status</b> (park ranking, national ranking, global ranking, international Red List ranking, nativeness), <b>and retained 2021 NPSpecies data</b> (alternative species name, old species occurrence designation, notes). These 22 features give a fairly detailed picture of amphibian presence and significance in different habitats across the country.  <p>
The authors confirmed taxonomy and conservation information from Integrated Taxonomic Information systems (ITIS), the International Union for Conservation of Nature (IUCN), and NatureServe.
#### Research Questions
The authors wanted to know if NPS lands contribute to the conservation of important amphibian populations. More specifically, they hypothesized that although most of the managed areas were not created to protect at-risk species, the geographic range of all the park units would support amphibian diversity, including protecting vulnerable and endangered species. The dataset presented in this paper was appropriate to answer this question given its verified park location, species identification, and conservation status information. <p>
Another question that could be asked using this data would be which NPS land unit provides sanctuary to the greatest number of endangered amphibians? Because this dataset provides park and conservation status information for each recorded species, it would be easy to determine which NPS site is home to the most at risk of extinction amphibians. Other researchers and conservationists could then use this information to better study and protect global amphibian diversity.

## Section 3: Visualizations
#### IUCNRank
Having used the `.describe()` function, we know that 3,270 out of 3,406 of the observations with "IUCNRank" data fall under "LC" or Least Concern. If we are only interested in at-risk species, defined as Near Threatened, Enangered, or Critically Endangered, we can create a new data frame that only contains theses species. Then we can ask a similar question to the one I posed in Part 2, Section 2, "Another Figure" which is: are certain amphibian taxonomic families more frequently at-risk than others? This dataset is appropriate to answer this question because it contains IUCN conservation statuses as well as taxonomy classifications. This visualization may provide insight into if certain amphibian families are more at-risk than others, which could help scientists prioritize protecting their populations. 

```python
# Find unique "IUCNRank" values
amphibians["IUCNRank"].unique()
```

```python
# Create a list of at-risk species IUCN conservation abbreviations
IUCN = ["NT","VU","EN","CR"]
# Create a boolean series where if the species is at-risk it returns True, else it returns False
IUCN_bool = amphibians["IUCNRank"].isin(IUCN)
# Create a new data frame that uses the boolean as an index so only at-risk species are included
IUCN_amph = amphibians.loc[IUCN_bool,:]
```

```python
# Using this new data frame of at-risk species, create a histogram
sns.displot(data=IUCN_amph, x="IUCNRank", hue="Family", multiple="stack")
#### The x-axis is the IUCN conservation status
#### The hues represent the different taxonomic families
#### The parameter multiple="stack" stacks the hues rather than overlaying them
```

This visualiztion shows us a few things:
1. Plethodonitidae are the most commonly observed at-risk family of amphibians in US National Park Services (NPS) land
2. Bufonidae are of particular concern given that at least one species is "Critically Endangered" and a few more are "Endangered"


These interpretations build off of Table 1 from the paper. This figure gives clarity as to which families of amphibians are most at-risk and of these, which are most observed across all NPS land. The table just gives high-level numbers of unqiue at-risk species observed.
#### CASC_Region
I now want to look at how geography and conservation status might be related. To do this, I will use the same at-risk species data frame created above for the "IUCNRank" visualization. This CASC_Region feature neatly classifies each observation by region which simplifies the US map such that we can more readily see trends based on location since we don't necessarily care if one observation is in northern Colorado and the other is in southern Colorado. The results can help government land managers know which regions to prioritize protecting if they want to most efficiently help at-risk species. 

```python
# Using the same at-risk spcies data frame from above, create a histogram
sns.displot(data=IUCN_amph, x="CASC_Region", hue="IUCNRank", multiple="stack")
#### The x-axis is the CASC_Region
#### The hues represent the IUCN conservation status
#### The parameter multiple="stack" stacks the hues rather than overlaying them
```

So while the `.describe()` function said that the most frequent CASC_Region for the whole amphibian dataset was the Northeast (NE) here we see that when only considering observations with an IUCNRank of Near Threatened or worse, the most frequently represented region is actually the Southwest (SW) followed by the Southeast (SE). This is an interesting find, and suggests that the coastal southern US National Park lands are habitat to the most at-risk amphibians. <p>
This interpretation builds off of Figure 2 of the paper, as it similarly looks at geographic spread of the data but in terms of conservation status rather than species diversity. In Figure 2, the Southeast shows high diversity whereas the Southwest shows only moderate diversity. However, with this visualization we see that both the Southeast and Southweest have high levels of at-risk species.
